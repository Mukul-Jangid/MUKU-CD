```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
dataset= pd.read_csv('Social_Network_Ads.csv')
dataset
```

```python
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, 4].values
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
print(X_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print('Confusion matrix: ')
print(cm)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % (accuracy_score(y_test, y_pred)*100))
print('Error: %.2f' % ((1-accuracy_score(y_test, y_pred))*100))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfmodel = RandomForestClassifier(n_estimators = 20, criterion ='entropy', random_state = 0)
rfmodel.fit(X_train, y_train) 
rf_pred = rfmodel.predict(X_test)
rf_pred
```

```python
from sklearn.metrics import confusion_matrix
cmc = confusion_matrix(y_test, rf_pred)
print('Confusion matrix: ')
print(cmc)
```

```python
from sklearn.metrics import accuracy_score
print('Accuracy: %.2f' % (accuracy_score(y_test, rf_pred)*100))
print('Error: %.2f' % ((1-accuracy_score(y_test, rf_pred))*100))
```
